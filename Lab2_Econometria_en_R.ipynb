```r
library(lubridate)
library(dplyr)
library(ggplot2)
```

```r
data <- read.csv("dataset.csv")
```

1. Cree un conjunto de columnas nuvas: día, mes, año, hora y minutos


```r
datos <- data %>%
  mutate(anio = year(dteday),
         mes = month(dteday),
         dia = day(dteday),
         hora = hour(dteday),
         minuto = minute(dteday))
```

```r
datos
```

2. ¿Qué mes es el que tiene la mayor demanda? Muestre una tabla y una
gráfica

```r
mes_mayor_demanda <- datos %>%
  group_by(mnth) %>%
  summarise(suma_mes = sum(cnt))
mes_mayor_demanda
```

```r
ggplot(mes_mayor_demanda, aes(x = mnth, y = suma_mes, fill = suma_mes)) +
  geom_bar(stat = 'identity',  col = 'black', lwd = 0)
```

El mes con mayor demanda es agosto.

3. ¿Qué rango de hora es la de mayor demanda? Muestre una tabla y una gráfica


```r
hora_mayor_demanda <- datos %>%
  group_by(hr) %>%
  summarise(suma_horas = sum(cnt))
hora_mayor_demanda
ggplot(hora_mayor_demanda, aes(x = hr, y = suma_horas)) +
  geom_bar(stat = 'identity', col = 'black', lwd = 1, fill = 'red')
```

El horario de mayor demanda es de 17 a 18 horas

4. ¿Qué temporada es la mas alta? Muestre una tabla


```r
temporada_mayor <- datos %>%
  group_by(season) %>%
  summarise(suma_temporadas = sum(cnt))
temporada_mayor

ggplot(temporada_mayor, aes(x = season, y = suma_temporadas)) +
  geom_bar(stat = 'identity', col = 'black', lwd = 1, fill = 'green')
```

 5. ¿A que temperatura disminuye la demanda? Muestre una gráfica para  analizar y dar su respuesta


```r
temperatura <- datos %>%
  group_by(temp)%>%
  summarise(suma_temperatura = sum(cnt), renta_prom = mean(cnt))
temperatura

ggplot(data = temperatura, aes(x = temp, y = suma_temperatura))+
  geom_bar(stat = 'identity', col = 'black', lwd = 1, fill = 'cyan')

ggplot(data = temperatura, aes(x = temp, y = renta_prom))+
  geom_line(stat = 'identity', col = 'blue', lwd = 1)
```

La demanda disminuye poco despues del 75% de la escala de temperatura, tiene un pico o maximo cerca del 100% y luego cae en picada

6. ¿A que humedad disminuye la demanda? Muestre una gráfica para  analizar y dar su respuesta.


```r
humedad <- datos %>%
  group_by(hum)%>%
  summarise(renta_humedad = sum(cnt), prom_humedad = mean(cnt))
humedad

ggplot(data = humedad, aes(x = hum, y = renta_humedad))+
  geom_bar(stat = 'identity', col = 'black', lwd = 0, fill = 'orange')

ggplot(data = humedad, aes(x = hum, y = prom_humedad))+
  geom_line(stat = 'identity', col = 'orange', lwd = 1)
```

La demanda empieza a disminuir cuando la humedad se encuentra cercana al 25%
